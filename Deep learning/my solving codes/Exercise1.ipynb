```python

import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

df=pd.read_csv('..\data\weight-height.csv')
df.plot(kind='scatter', x='Height',y='Weight', alpha=0.3, title='Male and Femal')
```

```python
df['Gender'].value_counts()
male=df[df['Gender']=='Male']
female=df[df['Gender']=='Female']
fig,ax=plt.subplots()
male.plot(kind='scatter',x='Height',y='Weight', c= 'r', ax=ax, alpha=0.3, title='Male')
female.plot(kind='scatter',x='Height',y='Weight', c= 'b',alpha=0.3, title='Female')
```

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

df=pd.read_csv('..\data\weight-height.csv')
male=df[df['Gender']=='Male']
female=df[df['Gender']=='Female']
fig,ax=plt.subplots()
male.plot(kind='hist', x='Height', ax=ax, alpha=0.5,color='r', title='Male ')  ### ax indicate for subplot
female.plot(kind='hist', x='Height',ax=ax, alpha=0.5,color='b', title=' Female')

plt.legend(["males", "females"])  ##### Lagen indicate data marking
plt.xlabel("Heigth (in)")

plt.axvline(male['Height'].mean(), color='red', linewidth=2)  #### draw  a vertical line at the mean of each population 
```

```python
import pandas as pd

import matplotlib.pyplot as plt

df=pd.read_csv('..\data\weight-height.csv')
dfpvt=df.pivot( columns='Gender', values='Weight')
dfpvt.head()
dfpvt.plot(kind='box')
plt.title('Weight Box Plot')
plt.xlabel("Weight (numbe)")           ### x axis label
plt.ylabel("Weight (lbs)");         ### y axis label 
```
