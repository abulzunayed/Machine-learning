```python
# Excercise 2
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('../data/weight-height.csv')
df.head()
#_=df.plot(kind='scatter', x='Height', y='Weight')
#plt.scatter(x=df['Height'], y=df['Weight'], c='r', alpha=0.3,cmap='viridis')
df['Gender'].value_counts()
male=df[df['Gender']=='Male']
female=df[df['Gender']=='Female']
fig,ax=plt.subplots()
male.plot(kind='scatter', x='Height', y='Weight', color='r',ax=ax, alpha=0.3, title='male')
female.plot(kind='scatter', x='Height', y='Weight', color='b', alpha=0.3, title='female')
plt.xlabel('Height-F')
plt.ylabel('Weight-M')
```

```python
#Exercise 3
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('../data/weight-height.csv')
female=df[df['Gender']=='Female']
fig,ax=plt.subplots()
female.plot(kind='hist', x='Height', bins=50, ax=ax, color='r', alpha=0.3,title ='female')
male.plot(kind='hist', x='Height', bins=50, ax=ax,color='b', alpha=0.4,title ='male')
plt.axvline(male['Height'].mean(), color='blue', linewidth=2)
plt.axvline(female['Height'].mean(), color='red', linewidth=2);
```

```python
#Exercise 4

import pandas as pd

import matplotlib.pyplot as plt

df=pd.read_csv('..\data\weight-height.csv')
dfpvt=df.pivot( columns='Gender', values='Weight')
dfpvt.head()
dfpvt.plot(kind='box')
plt.title('Weight Box Plot')
plt.xlabel("Weight (numbe)")           ### x axis label
plt.ylabel("Weight (lbs)"); 
```
